```python
import torch
import os
import matplotlib.pyplot as plt
import numpy as np

from PIL import Image
from torchvision import transforms
from torch.utils.data import DataLoader
from tqdm import tqdm
import pandas as pd
from Examples.ModelNet10.dataset import ModelNet10
```

```python
root = '../Datasets/ModelNet10_pics/'
train_set = ModelNet10(root, 'train', n=1)
val_set = ModelNet10(root, 'val', n=1)
```

```python
fig, axes = plt.subplots(2, 10, figsize=(20, 4))
for i in range(10):
    axes[0, i].imshow(train_set[i][0][0].permute(1, 2, 0))
    axes[0, i].axis('off')
    axes[0, i].set_title(f'Image {i+1}')
    
    axes[1, i].imshow(val_set[i][0][0].permute(1, 2, 0))
    axes[1, i].axis('off')
    axes[1, i].set_title(f'Image2 {i+1}')

plt.show()
```
